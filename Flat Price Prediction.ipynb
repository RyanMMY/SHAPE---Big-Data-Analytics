# Data Preprocessing 

```python
import pandas as pd

df = pd.read_csv('DBA_1-4.csv', encoding='utf-8')
```

```python
df
```

```python
# Draw the floor information from column 'b'
df['Floor'] = df['b'].str.split(' ').str[0]
df['Floor'] = df['Floor'].replace({'高層':'3', '中層':'2', '低層':'1'}, regex = True)
```

```python
# Draw the building year information from column 'd'
df['Building_Year'] = df['d'].str.split('年樓齡').str[0]
```

```python
# Draw the saleable area information from column 'e'
df['Saleable_Area'] = df['e'].str.split('呎').str[0]
```

```python
# Draw the flats sale price information from column 'e'
df['Flats_Price'] = df['f'].str.split('\n').str[0]
df['Flats_Price'] = df['Flats_Price'].str.split('售').str[1]
df['Flats_Price'] = df['Flats_Price'].replace({'\$':'', '萬':'', ',':''}, regex = True)
df['Flats_Price'] = df['Flats_Price'].astype('str')

Million = "0000"
df['Flats_Price'] = [x + Million for x in df['Flats_Price']]
```

```python
# Draw the flats room number information from column 'a'
df['Number_of_Room'] = df['a'].str.split('房').str[0]
df['Number_of_Room'] = df['Number_of_Room'].str.split().str[-1]
```

```python
# Draw the building name from column 'a'
df['Building_Name'] = df['a'].str.split().str[0]
```

# Data Cleaning

```python
# Delete all space and '\n' in the dataframe
df = df.replace({'\n':'', ',':'', ' ':'', '·':'', '%':''}, regex = True)
```

```python
# Drop the useless columns from the dataframe
df = df.drop(columns=['Unnamed: 0', 'a', 'b', 'd', 'e', 'f'])
```

```python
# Rename the columns
df.rename(columns = {'c':'Location', 'g':'District'}, inplace = True)
```

```python
# Delete all unnecessary string in the columns
df = df.drop(df[df.Number_of_Room.str.contains(r'[a-zA-Z, 室]')].index)
df = df.drop(df[df.Building_Year.str.contains(r'%')].index)
df = df.drop(df[df.Flats_Price.str.contains(r'nan')].index)
```

```python
# Drop rows if the value in columns is not an integer
df = df[df['Building_Year'].apply(lambda x: str(x).isdigit())]
df = df[df['Flats_Price'].apply(lambda x: str(x).isdigit())]
df = df[df['Number_of_Room'].apply(lambda x: str(x).isdigit())]
df = df[df['Floor'].apply(lambda x: str(x).isdigit())]
```

```python
# Change the data type
df['Building_Year'] = df['Building_Year'].astype('int64')
df['Saleable_Area'] = df['Saleable_Area'].astype('int64')
df['Flats_Price'] = df['Flats_Price'].astype('int64')
df['Number_of_Room'] = df['Number_of_Room'].astype('int64')
df['Floor'] = df['Floor'].astype('int64')
```

```python
df = df.reset_index(drop=True)
```

# Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

cols = ['District','Floor','Building_Year','Saleable_Area','Number_of_Room']
X = df[cols].values
y = df['Flats_Price'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

### Linear Regression

```python
# Create Linear regressor
from sklearn.linear_model import LinearRegression
import numpy as np

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
```

```python
# Train Set Evaluation
print('Linear Regression Train Set Evaluation: \n--------------------------------------------------')
print('R^2:', metrics.r2_score(y_train, y_pred))
print('Adjusted R^2:', 1 - (1-metrics.r2_score(y_train, y_pred))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
print('Mean of absolute error:', metrics.mean_absolute_error(y_train, y_pred))
print('Mean-square error:', metrics.mean_squared_error(y_train, y_pred))
print('Root mean-square error:', np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```

```python
# Difference between actual price and predicted value
plt.scatter(y_train, y_pred)
diagonal = np.linspace(0, np.max(y_test), 100)
plt.plot(diagonal, diagonal, '-r')
plt.title("Prices vs Predicted Prices(Linear Regression)")
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.show()
```

```python
y_test_pred = lr.predict(X_test)
```

```python
# Test Set Evaluation
acc_linreg = metrics.r2_score(y_test, y_test_pred)
print('Linear Regression Test Set Evaluation: \n--------------------------------------------------')
print('R^2:', acc_linreg)
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_test, y_test_pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
print('Mean of absolute error:',metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean-square error:',metrics.mean_squared_error(y_test, y_test_pred))
print('Root mean-square error:',np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

### Random Forest Regressor

```python
# Create Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor()

reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_train)
```

```python
# Train Set Evaluation
print('Random Forest Regressor Train Set Evaluation: \n--------------------------------------------------')
print('R^2:',metrics.r2_score(y_train, y_pred))
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_train, y_pred))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
print('Mean of absolute error:',metrics.mean_absolute_error(y_train, y_pred))
print('Mean-square error:',metrics.mean_squared_error(y_train, y_pred))
print('Root mean-square error:',np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```

```python
# Difference between actual price and predicted value
plt.scatter(y_train, y_pred)
diagonal = np.linspace(0, np.max(y_test), 100)
plt.plot(diagonal, diagonal, '-r')
plt.title("Prices vs Predicted Prices(Random Forest Regressor)")
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.show()
```

```python
y_test_pred = reg.predict(X_test)
```

```python
# Test Set Evaluation
acc_rf = metrics.r2_score(y_test, y_test_pred)
print('Random Forest Regressor Test Set Evaluation: \n--------------------------------------------------')
print('R^2:',acc_rf)
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_test, y_test_pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
print('Mean of absolute error:',metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean-square error:',metrics.mean_squared_error(y_test, y_test_pred))
print('Root mean-square error:',np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```
